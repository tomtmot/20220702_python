```python
def draw_map(place):
  gps=geocoder.osm(place).latlng
  
  m=folium.Map(location=gps, zoom_start=16)
  folium.Marker(gps, popup=place, icon=folium.Icon(color="red")).add_to(m)
 

  return m
```
